- Exercise 1  
 $f(x)=sin(x)+cos(x)$  
<font color="green">$f(x)=sin(x)+cos(x)$ </font>

```python
import matplotlib.pyplot as plt
import numpy as np
a = 50
c = ['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$', '2$\pi$', '3$\pi$']
b = np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi
x = np.linspace(-3*np.pi, 3*np.pi, a)
y = np.sin(x)+np.cos(x)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots(figsize=(6, 4), facecolor='#F5DEB3')
plt.plot(x, y, color='#40E0D0', label='sin(x)+cos(x)')
plt.plot(x, y1, color='#DDA0DD', linestyle='--')
plt.plot(x, y2, color='#800000', linestyle='-.')
plt.grid(True, color='#DEB26F')
# ax.patch.set_facecolor("#F5DEB3")
plt.xlabel('X', fontsize='x-large')
plt.xticks(b, c)
plt.ylim(-3, 3)
plt.ylabel('Y', fontsize='x-large')
plt.axhline(2**(1/2), c='#888888', ls='--', label='$y = \sqrt{2}$')  # 漸 進 線
plt.axhline(-(2**(1/2)), c='#888888',
            ls=(0, (3, 5, 1, 5)), label='$y = -\sqrt{2}$')
plt.title("$sin(x)+cos(x)$", fontsize=13)
plt.legend()
plt.savefig(air+'/hw2ex1.pdf', format='pdf')
plt.show()
```

- Exercise 2 $f(x)=\frac{1-e^{-2x}}{1+e^{-2x}}$

```python
import matplotlib.pyplot as plt
import numpy as np
b = 0.5
x_max = 10
x_min = -10
x = np.arange(x_min, x_max, b)
# formulate a function f
def f(x): return (1-np.exp(-2*x))/(1+np.exp(-2*x))
fig, ax = plt.subplots(figsize=(6, 4), facecolor='#F5F5DC')
plt.plot(x, f(x), linestyle=':', color='#ADD8E6', marker='$♥$',
         markersize=10, label=r'$f\ (x)=\frac{1-e^{-2x}}{1+e^{-2x}}$')
plt.grid(visible=True, color='#00FFFF', linestyle='-.', linewidth=0.5)
plt.xlabel('X', fontsize='large', color='#4B0082')
plt.ylabel('Y', fontsize='large', color='#4B0082')
plt.ylim(-2, 2)
plt.title('$f(x)= \\frac{1-e^{-2x}}{1+e^{-2x}}$', fontsize=15, color='#4B0082')
plt.xticks(color='#4B0082'), plt.yticks(color='#4B0082')
plt.axhline(1, c='#888888', ls=(0, (3, 5, 1, 5, 1, 5)), label='$y=1$')
plt.axhline(-1, c='#888888', ls='--', label='$y = -1$')
ax.annotate('Inflection point', xy=(0, 0), xytext=(2, 0), arrowprops=dict(
    arrowstyle='->', connectionstyle='arc3, rad=0.3', color='#872657'), fontsize=11, color='#872657')
plt.legend(loc='upper left')
plt.savefig(air+'/hw2ex2.pdf', format='pdf')
plt.show()
```

- Exercise 5 $f(x)=\frac{1}{2\sqrt{2\pi}}e^{-\frac{(x-1)^2}{8}}$

```python
import matplotlib.pyplot as plt
import numpy as np
xmin, xmax = -10, 10
x = np.linspace(xmin, xmax, 100)
y = (1/(2*np.sqrt(2*np.pi)))*np.exp(-((x-1)**2)/8)
fig = plt.figure()
# use axis object
ax = plt.gca()  # get current axis
ax.patch.set_facecolor("gray")
ax.plot(x, y, linewidth=3, color='purple', linestyle='-.')
ax.grid(True, color='orange')
ax.set_xlabel('X'), ax.set_ylabel('f(x)')
ax.set_title('$f(x)=\\frac{1}{2\sqrt{2\pi}}e^{-\\frac{(x-1)^2}{8}}$')
# plt.savefig('ex5.jpg', format='jpg')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
xmin, xmax = -10, 10
x = np.linspace(xmin, xmax, 100)
y = (1/(2*np.sqrt(2*np.pi)))*np.exp(-((x-1)**2)/8)
dydx = np.cos(0.5 * (x[:-1] + x[1:]))  # first derivative
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
fig, axs = plt.subplots(facecolor='#F5F5DC')
# Create a continuous norm to map from data points to colors
norm = plt.Normalize(dydx.min(), dydx.max())
lc = LineCollection(segments, cmap='viridis', norm=norm)
# Set the values used for colormapping
lc.set_array(dydx)
lc.set_linewidth(2)
line = axs.add_collection(lc)
fig.colorbar(line, ax=axs)
# Use a boundary norm instead
cmap = ListedColormap(['r', 'g', 'b'])
norm = BoundaryNorm([-1, -0.5, 0.5, 1], cmap.N)
lc = LineCollection(segments, cmap=cmap, norm=norm)
lc.set_array(dydx)
lc.set_linewidth(2)
axs.grid()
axs.set_title('$f(x)=\\frac{1}{2\sqrt{2\pi}}e^{-\\frac{(x-1)^2}{8}}$')
axs.set_xlim(x.min(), x.max())
axs.set_ylim(0, 0.25)
axs.set_xlabel('X'), axs.set_ylabel('f(x)')
axs.vlines(1, 0, 0.2, color='#87CEFF',linestyle='--', linewidth=2, label='$x=1$')
plt.savefig(air+'/hw2ex5.pdf', format='pdf')
plt.show()
```
