$\\Normal$

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import norm
import matplotlib.pyplot as plt

size=100000
rng=np.random.default_rng(5)
x=rng.normal(loc=0,scale=1,size=size)
fig,ax=plt.subplots(2,2,figsize=(12,10))
ax[0][0].hist(x,bins=30,alpha=0.6,color='b',linewidth = 1, density = True, rwidth = 0.9)

# x1 = norm.rvs(size = size, random_state = rng)
boxprops = dict(linestyle = '--', linewidth = 3, color = 'darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor = 'green', 
    markersize = 8, linestyle = 'none') # define outliers
labels = ['N(0,1)']
ax[0][1].set_yticks(np.arange(-4,5.0,1))
ax[0][1].grid(True)
# ax[0][1].set_ylim(-2,2)
ax[0][1].boxplot(x, boxprops = boxprops,
    flierprops = flierprops, labels = labels)

# x2=np.random.normal(loc=0,scale=1,size=size)    
stats.probplot(x, dist = "norm", plot = ax[1][0])

# sample=norm.rvs(size=size)
x1=np.sort(x)
y=np.arange(1,size+1)/size
ax[1][1].plot(x1,y,drawstyle='steps-pre',label='Emperical CDF')
y_=norm.cdf(x1)
ax[1][1].plot(x1,y_,color='r', linestyle = '--',
    linewidth = 5, alpha = 0.5, label = 'real CDF')
ax[1][1].legend()
plt.suptitle('$Normal(0,1)$',fontsize=15)
plt.grid(True)    
# plt.savefig('hw3normal4.png', format='png')
plt.show()
```

至少 3 種以上的抽樣分配。也就是 Y=g(X1,X2,...,Xn) 不同的 g(.) 函數。
最好都能找到（畫出）其理論上的抽樣分配。
至少一種來自離散型的變數。
抽樣分配與樣本數有密切關係，必須呈現出樣本數對抽樣分配造成的影響。
以適當的圖形來表達實驗結果，譬如直方圖、 empirical cdf 圖。

$Y=X^2$ $\\$ 其中X~$\\Norm(0,1)$

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2,poisson,f,norm
import matplotlib.pyplot as plt
size=100000
rng=np.random.default_rng(4)
x=rng.normal(loc=0,scale=1,size=size)
w=x**2

df=1
fig,ax=plt.subplots(2,2,figsize=(12,10))
ax[0][0].hist(w,bins=50,alpha=0.6,color='b',linewidth = 1, density=True, rwidth = 0.9)
xlim=[0,10]
x_=np.linspace(xlim[0],xlim[1],500)
y_chi2=chi2.pdf(x_.reshape(-1,1),df)
ax[0][0].plot(x_,y_chi2,color='r')
ax[0][0].set_xlim(0,10)

boxprops = dict(linestyle = '--', linewidth = 3, color = 'darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor = 'green', 
    markersize = 8, linestyle = 'none') # define outliers
labels = ['$X^2$']
ax[0][1].boxplot(w, boxprops = boxprops,
    flierprops = flierprops, labels = labels)

stats.probplot(w, dist = "norm", plot = ax[1][0])

x1=np.sort(w)
y=np.arange(1,size+1)/size
ax[1][1].plot(x1,y,drawstyle='steps-pre',label='Emperical CDF')
y_=norm.cdf(x1)
ax[1][1].plot(x1,y_,color='r', linestyle = '--',
    linewidth = 5, alpha = 0.5, label = 'real CDF')
ax[1][1].legend()
ax[1][1].set_facecolor("yellow")
fig.set_facecolor("violet")
plt.suptitle('Sampling distribution of $Y=X^2$,n=100000',fontsize=15)
plt.grid(True)  
plt.savefig(img_dir+'hw3sumstdnor.png', format='png')  
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2,poisson,f,norm,bernoulli,binom
import matplotlib.pyplot as plt
size=10000
rng=np.random.default_rng()
x1=rng.binomial(p=0.5,n=100,size=size)
x2=rng.binomial(p=0.5,n=100,size=size)
x3=rng.binomial(p=0.5,n=100,size=size)
x4=rng.binomial(p=0.5,n=100,size=size)
w=x1+x2+x3+x4

fig,ax=plt.subplots(2,2,figsize=(12,10))
ax[0][0].hist(w,bins=50,alpha=0.4,color='b',linewidth = 1, density=True, rwidth = 0.9)
x_=np.linspace(0,500)
y_bin=binom.pmf(x_.reshape(-1,1),n=400,p=0.5)
ax[0][0].stem(x_,y_bin)
ax[0][0].set_xlim(100,300)

boxprops = dict(linestyle = '--', linewidth = 3, color = 'darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor = 'green', 
    markersize = 8, linestyle = 'none') # define outliers
labels = ['$Y$']
ax[0][1].boxplot(w, boxprops = boxprops,
    flierprops = flierprops, labels = labels)

stats.probplot(w, dist = "norm", plot = ax[1][0])

x1=np.sort(w)
y=np.arange(1,size+1)/size
ax[1][1].plot(x1,y,drawstyle='steps-pre',label='Emperical CDF')
y_=binom.cdf(x1,n=400,p=0.5)
ax[1][1].plot(x1,y_,color='r', linestyle = '--',
    linewidth = 5, alpha = 0.5, label = 'real CDF')
ax[1][1].legend()
plt.suptitle('Sampling distribution of $Y=X_1+x_2+x_3+x_4$,n=10000',fontsize=15)
plt.grid(True)
plt.savefig(img_dir+'hw3sumbernoulli.png', format='png')    
plt.show()
```

給定四個數字 (2, 4, 9, 12)。從這四個數字中隨機抽取四個數字（取後放回）並計算其平均數。假設隨機變數 Y 代表這四個數字的平均數。請繪製隨機變數 Y 的 PMF。本題可以直接計算每個平均數的機率，但在此請使用隨機抽樣的方式，估計出這些機率值。其中抽樣的次數可以高至百萬以上。
Hint: 取後放回的抽樣方式，可以使用 numpy 的 random.choice。而從重複的數字中取出不重複的部分，可以使用 numpy 的 unique。（本題出自 Statistical Inference by G. Casella & R.L. Berger）

```python
import numpy as np
import matplotlib.pyplot as plt
#method1:loop
x=np.zeros(1000000)
for i in range(len(x)):
    a=np.random.choice([2,4,9,12],size=4, replace=True)
    y_bar=np.mean(a)
    x[i]=y_bar
plt.hist(x,bins=100,density='True',rwidth = 0.7)
plt.show()
plt.savefig('hw3samplingloop.png', format='png')
#method2:matrix
a=[2,4,9,12]
n=1000000
sum=np.zeros(shape=n)
x=np.random.choice(a,size=(n,4))
x_bar=x.mean(axis=1)
weights=np.ones_like(x_bar)/len(x_bar)
plt.hist(x_bar,bins=40,alpha=0.6,color='b',edgecolor='y',weights=weights,rwidth=0.7)#density=True
plt.savefig(img_dir+'hw3samplingmatrix.png', format='png')
plt.show()
```
