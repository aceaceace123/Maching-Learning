```python
import numpy as np
import scipy.stats as stats
from scipy.stats import norm
import matplotlib.pyplot as plt

size=100000
rng=np.random.default_rng(5)
x=rng.normal(loc=0,scale=1,size=size)
fig,ax=plt.subplots(2,2,figsize=(12,10))
ax[0][0].hist(x,bins=30,alpha=0.6,color='b',linewidth = 1, density = True, rwidth = 0.9)

# x1 = norm.rvs(size = size, random_state = rng)
boxprops = dict(linestyle = '--', linewidth = 3, color = 'darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor = 'green', 
    markersize = 8, linestyle = 'none') # define outliers
labels = ['N(0,1)']
ax[0][1].set_yticks(np.arange(-4,5.0,1))
ax[0][1].grid(True)
# ax[0][1].set_ylim(-2,2)
ax[0][1].boxplot(x, boxprops = boxprops,
    flierprops = flierprops, labels = labels)

# x2=np.random.normal(loc=0,scale=1,size=size)    
stats.probplot(x, dist = "norm", plot = ax[1][0])

# sample=norm.rvs(size=size)
x1=np.sort(x)
y=np.arange(1,size+1)/size
ax[1][1].plot(x1,y,drawstyle='steps-pre',label='Emperical CDF')
y_=norm.cdf(x1)
ax[1][1].plot(x1,y_,color='r', linestyle = '--',
    linewidth = 5, alpha = 0.5, label = 'real CDF')
ax[1][1].legend()
plt.suptitle('$Normal(0,1)$',fontsize=15)
plt.grid(True)    
plt.show()
plt.savefig('hw3normal4.png', format='png')
```

$Y=X^2$ $\\$ X~$\\Norm(0,1)$

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2,poisson,f,norm
import matplotlib.pyplot as plt
size=100000
rng=np.random.default_rng(4)
x=rng.normal(loc=0,scale=1,size=size)
w=x**2

df=1
fig,ax=plt.subplots(2,2,figsize=(12,10))
ax[0][0].hist(w,bins=50,alpha=0.6,color='b',linewidth = 1, density=True, rwidth = 0.9)
xlim=[0,10]
x_=np.linspace(xlim[0],xlim[1],500)
y_chi2=chi2.pdf(x_.reshape(-1,1),df)
ax[0][0].plot(x_,y_chi2,color='r')
ax[0][0].set_xlim(0,10)

boxprops = dict(linestyle = '--', linewidth = 3, color = 'darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor = 'green', 
    markersize = 8, linestyle = 'none') # define outliers
labels = ['$X^2$']
ax[0][1].boxplot(w, boxprops = boxprops,
    flierprops = flierprops, labels = labels)

stats.probplot(w, dist = "norm", plot = ax[1][0])

x1=np.sort(w)
y=np.arange(1,size+1)/size
ax[1][1].plot(x1,y,drawstyle='steps-pre',label='Emperical CDF')
y_=norm.cdf(x1)
ax[1][1].plot(x1,y_,color='r', linestyle = '--',
    linewidth = 5, alpha = 0.5, label = 'real CDF')
ax[1][1].legend()
plt.suptitle('Sampling distribution of $Y=X^2$,n=100000',fontsize=15)
plt.grid(True)    
plt.show()
plt.savefig('hw3sumstdnor.png', format='png')
```
